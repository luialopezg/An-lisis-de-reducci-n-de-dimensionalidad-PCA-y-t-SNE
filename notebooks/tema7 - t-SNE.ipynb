# Ejemplo con t-SNE

## 1. Dataset Iris

Variables
* Sepal Length
* Sepal Width
* Petal Length
* Petal Width

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.manifold import TSNE
from sklearn.preprocessing import StandardScaler
```

```python
# Cargar el conjunto de datos Iris
iris = datasets.load_iris()
X = iris.data
y = iris.target
target_names = iris.target_names

# Estandarizar las características
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
display(pd.DataFrame(X, columns=iris.feature_names).head())
display(pd.DataFrame(X_scaled, columns=iris.feature_names).head())
```

```python
# Aplicar t-SNE para reducir las dimensiones a 2D
tsne = TSNE(n_components=2, perplexity=30, n_iter=1000, random_state=42)
X_tsne = tsne.fit_transform(X_scaled)

# Convertir a DataFrame para visualización más fácil
df_tsne = pd.DataFrame(X_tsne, columns=['Dim1', 'Dim2'])
df_tsne['target'] = y
```

```python
# Visualización de los resultados con matplotlib
plt.figure(figsize=(8, 6))
colors = ['r', 'g', 'b']

for i, target_name in enumerate(target_names):
    plt.scatter(df_tsne.loc[df_tsne['target'] == i, 'Dim1'],
                df_tsne.loc[df_tsne['target'] == i, 'Dim2'],
                label=target_name, color=colors[i])

plt.legend()
plt.title('t-SNE en el conjunto de datos Iris')
plt.xlabel('Dim1')
plt.ylabel('Dim2')
plt.show()
```

## 2. Dataset vinos

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.manifold import TSNE
from sklearn.preprocessing import StandardScaler
```

```python
# Cargar el conjunto de datos de vinos
wine = datasets.load_wine()
X = wine.data
y = wine.target
target_names = wine.target_names

# Estandarizar las características (es importante para t-SNE)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
display(pd.DataFrame(X, columns=wine.feature_names).head())
display(pd.DataFrame(X_scaled, columns=wine.feature_names).head())
```

```python
# Aplicar t-SNE para reducir las dimensiones a 2D
tsne = TSNE(n_components=2, perplexity=30, n_iter=1000, random_state=42)
X_tsne = tsne.fit_transform(X_scaled)

# Convertir a DataFrame para visualización más fácil
df_tsne = pd.DataFrame(X_tsne, columns=['Dim1', 'Dim2'])
df_tsne['target'] = y
```

```python
# Visualización de los resultados con matplotlib
plt.figure(figsize=(8, 6))
colors = ['r', 'g', 'b']

for i, target_name in enumerate(target_names):
    plt.scatter(df_tsne.loc[df_tsne['target'] == i, 'Dim1'],
                df_tsne.loc[df_tsne['target'] == i, 'Dim2'],
                label=target_name, color=colors[i])

plt.legend()
plt.title('t-SNE en el conjunto de datos vinos')
plt.xlabel('Dim1')
plt.ylabel('Dim2')
plt.show()
```
